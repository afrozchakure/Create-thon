```python
# Loading our file into a data frame
df = pd.read_excel('Data_Train.xlsx')   # For the training set
df2 = pd.read_excel('Data_Test.xlsx')   # For the testing set
```

```python
# Cleaning our Train data
df['Engine'] = df['Engine'].str.replace('CC', '')
df['Power'] = df['Power'].str.replace('bhp', '')
df['Mileage'] = df['Mileage'].str.replace('km/kg', '')
df['Mileage'] = df['Mileage'].str.replace('kmpl', '')
df['Power'] = df['Power'].str.replace('null', 'NaN')
df['New_Price'] = df['New_Price'].str.replace('Lakh', '') # New_Price

# Cleaning our Test data
df2['Engine'] = df2['Engine'].str.replace('CC', '')
df2['Power'] = df2['Power'].str.replace('bhp', '')
df2['Mileage'] = df2['Mileage'].str.replace('km/kg', '')
df2['Mileage'] = df2['Mileage'].str.replace('kmpl', '')
df2['Power'] = df2['Power'].str.replace('null', 'NaN')
df2['New_Price'] = df2['New_Price'].str.replace('Lakh','')  # New_Price
df2.head()
# df.Mileage = [x.strip('km/kg') for x in df.Mileage]
# dataset.info()
df2.head()
```

```python
# Splitting our data into dependent and independent variables for training set
y = df.iloc[:, 10].values
X = df.iloc[:, :10].values
df.drop('Price', axis = 1, inplace = True)
df.head()

X2 = df2.iloc[:, :10].values  # Testing Data assigned to dependent variable
```

```python
# Mark empty values as missing or NaN
df[['Mileage', 'Engine', 'Power', 'Seats']] = df[['Mileage','Engine','Power', 'Seats']].replace('', np.NaN)
df2[['Mileage', 'Engine', 'Power', 'Seats']] = df2[['Mileage','Engine','Power', 'Seats']].replace('', np.NaN)
```

```python
df['train'] = 1
df2['train'] = 0
combined = pd.concat((df, df2))  # Combining the DataFrames
s1 = ['Fuel_Type', 'Transmission', 'Owner_Type', 'Location']
new_df = pd.get_dummies(combined[s1])
# new_df = pd.get_dummies(combined['Fuel_Type'])
# new_df = pd.get_dummies(combined['Transmission'])
# new_df = pd.get_dummies(combined['Owner_Type'])
```

```python
new_df.head()
# combined.head()
```

```python
combined = pd.concat([combined, new_df], axis = 1)
train_df = combined[combined['train'] == 1]
test_df = combined[combined['train'] == 0]
s1 = ['Fuel_Type', 'Transmission', 'Owner_Type', 'Location', 'train']
train_df.drop(s1, axis = 1, inplace = True)
test_df.drop(s1, axis = 1, inplace = True)
test_df.head()
```

```python
train_df.head()
X = train_df.iloc[:,:].values
X2 = test_df.iloc[:,:].values
```

```python
# Replacing missing values with the mean of the columns
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values = 'NaN', strategy = 'mean', axis = 0)
imputer = imputer.fit(X[:, 2:])
X[:, 2:] = imputer.transform(X[:, 2:])

imputer = imputer.fit(X2[:, 2:])
X2[:, 2:] = imputer.transform(X2[:, 2:])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators = 175, criterion = 'mse', random_state = 0)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn import metrics  
print("MAE", metrics.mean_absolute_error(y_test, y_pred))
print("MSE", metrics.mean_squared_error(y_test, y_pred))
print("RMSE", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print("Score:", model.score(X_test, y_test))
```
